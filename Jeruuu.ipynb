```python
# File: https://drive.google.com/file/d/14Z2jg-sL39QX1djAIqRh_TQZL5SXEZmc/view?usp=drive_link
import gdown
gdown.download("https://drive.google.com/uc?id=14Z2jg-sL39QX1djAIqRh_TQZL5SXEZmc", "Akhire_Lungo.mp3", quiet=True)

```

```python
display(Audio("Akhire_Lungo.mp3", autoplay=True))

lock = Lock()

def animate_text(text, delay=0.1):
    with lock:
        for char in text:
            sys.stdout.write(char)
            sys.stdout.flush()
            time.sleep(delay)
        print()

def sing_lyric(lyric, delay, speed):
    time.sleep(delay)
    animate_text(lyric, speed)

def sing_song():
    lyrics = [
        ("\nAkhire koowe....", 0.10),
        ("milih lungo ninggal aku...", 0.12),
        ("seng wes ngupayake", 0.10),
        ("ben ra dadi cerito lloorooooo...", 0.11),
        ("Ajure roso iki ngerti kowe lungoo....", 0.13),
        ("tak jaluk kowe...", 0.12),
        ("ojoo", 0.09),
        ("nganti geloo...\n", 0.11),
    ]

    delays = [2.2, 4.9, 9.1, 12.7, 19.9, 26.9, 29.7, 30.8]

    threads = []
    for i in range(len(lyrics)):
        lyric, speed = lyrics[i]
        t = Thread(target=sing_lyric, args=(lyric, delays[i], speed))
        threads.append(t)
        t.start()

    for thread in threads:
        thread.join()

sing_song()
```
